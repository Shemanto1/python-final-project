# *Number of airlines in the dataset and  their frequencies

```python
data['airline'].nunique()
# no. of airline
```

```python
data['airline'].unique()
#airline name
```

```python
# showing airlines with their frequiences..

data['airline'].value_counts(ascending =True)
```

```python
# showing airlines with frequencies using horizontal bar

data['airline'].value_counts(ascending = True).plot.barh(color =['red','lightgreen'])
plt.title("Airline with frequencies ")
plt.xlabel("no. of flights")
plt.ylabel("Airlines")
plt.show()
```

# *Showing  bar graphs representing departure time and arrival time

```python
data.head()
```

```python
#showing the departure time
data ['departure_time'].value_counts(ascending =True)
```

```python
#showing the arrival time
data ['arrival_time'].value_counts(ascending =True)
```

```python
# showing in bar (departure and arrival)

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(data ['departure_time'].value_counts(ascending =True).index,data ['departure_time'].value_counts(ascending =True).values,color =['red','blue'])

plt.xlabel("Departure Time")
plt.ylabel("Departure frequencies")

plt.subplot(1,2,2)
plt.bar(data ['arrival_time'].value_counts(ascending =True).index,data ['arrival_time'].value_counts(ascending =True).values,color =['green','orange'])

plt.xlabel("Arrival Time")
plt.ylabel("Arrival frequencies")

plt.show()
```
